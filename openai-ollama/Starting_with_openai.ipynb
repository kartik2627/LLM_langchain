🔥 Core Difference (One Line)

👉 Retriever = finds information
👉 Document Chain = uses that information to generate answer

🧠 1. Retriever (What it does)

In LangChain:

👉 A retriever searches your data and returns relevant documents.

📦 Flow
Question → Retriever → Relevant Docs
🔧 Example
retriever = vectorstore.as_retriever()
docs = retriever.invoke("What is LangChain?")

👉 Output:

List of Document objects
No answer yet ❌
🧠 Think of it like:

👨‍🔍 Google Search
👉 It gives links, not answers

✅ Responsibilities
Search vector DB
Perform similarity search
Return top-k documents
📄 2. Document Chain (What it does)

👉 A document chain takes those documents and generates the final answer using LLM.

📦 Flow
Docs → Prompt → LLM → Answer
🔧 Example
response = document_chain.invoke({
    "context": docs,
    "input": "What is LangChain?"
})

👉 Output:

Final answer ✅
🧠 Think of it like:

✍️ Writer
👉 Reads documents and writes answer

✅ Responsibilities
Combine documents
Insert into {context}
Call LLM
Generate answer

```python
# ================================
# 1. ENV SETUP
# ================================
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

# Optional (LangSmith tracking)
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = "RAG-Demo"

# ================================
# 2. IMPORTS
# ================================
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.documents import Document

from langchain_community.vectorstores import FAISS
from langchain_classic.chains.combine_documents.stuff import create_stuff_documents_chain
from langchain_classic.chains import create_retrieval_chain

# ================================
# 3. CREATE SAMPLE DOCUMENTS
# ================================
docs = [
    Document(page_content="LangChain is a framework for building LLM-powered applications."),
    Document(page_content="It supports RAG, agents, prompt templates, and chains."),
    Document(page_content="RAG stands for Retrieval-Augmented Generation."),
]

# ================================
# 4. CREATE EMBEDDINGS + VECTOR STORE
# ================================
embeddings = OpenAIEmbeddings()

vectorstore = FAISS.from_documents(docs, embeddings)

# ================================
# 5. CREATE RETRIEVER (RETRIEVAL CHAIN PART)
# ================================
retriever = vectorstore.as_retriever()

# ================================
# 6. CREATE LLM
# ================================
llm = ChatOpenAI(model="gpt-4o")

# ================================
# 7. CREATE PROMPT (IMPORTANT: {context})
# ================================
prompt = ChatPromptTemplate.from_template(
    """
Answer the question based only on the provided context.

<context>
{context}
</context>

Question: {input}
"""
)

# ================================
# 8. CREATE DOCUMENT CHAIN
# ================================
document_chain = create_stuff_documents_chain(llm, prompt)

# ================================
# 9. CREATE RETRIEVAL CHAIN (FULL RAG)
# ================================
rag_chain = create_retrieval_chain(retriever, document_chain)

# ================================
# 10. RUN QUERY
# ================================
response = rag_chain.invoke({
    "input": "What is LangChain?"
})

print(response["answer"])
```
